```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython import display
import numpy.ma as ma
```

```python
with open('data_s_graded.npy', 'rb') as f:
    data_s = np.load(f, allow_pickle=True).tolist()["Process_6"]["s_out"]
```

```python
np.argwhere(data_s)
```

```python
data_s.shape
```

```python
result = np.zeros((data_s[0].shape))
for x in range(data_s[0].shape[1]):
    for y in range(data_s[0].shape[0]):
        t_last = np.argwhere(data_s[:,y,x])
        if t_last.size == 0:
            continue
        result[y,x] = data_s[t_last.max(), y, x]
```

```python
plt.imshow(result, cmap="jet")
```

```python
test = np.exp(result)
plt.hist(result.flatten(), 100)
```

```python
m = ma.masked_where(np.logical_or(result>60, result == 0), result)
```

```python
plt.figure(figsize=(10, 10))
plt.imshow(m, cmap="jet")
```
