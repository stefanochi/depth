```python
import numpy as np
import matplotlib.pyplot as plt

path = "C:/Users/schiavaz/lava-nc/depth/"
dataset = "data/stripes/"
```

```python
events = np.loadtxt(path + dataset + "events.txt")
```

```python
events.shape
```

```python
events = events[100000:200000]
```

```python
xl = 50
xh = 70
yl = 90
yh = 110
```

```python
events = events[np.logical_and(events[:, 1] > xl, events[:, 1] < xh)]
```

```python
events = events[np.logical_and(events[:, 2] > yl, events[:, 2] < yh)]
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
%matplotlib notebook
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(events[:, 0], events[:, 1], events[:, 2])

ax.set_xlabel("time", fontsize=16)
ax.set_ylabel("x", fontsize=16)
ax.set_zlabel("y", fontsize=16)
```
