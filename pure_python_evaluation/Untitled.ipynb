```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_shape = (180, 180)
U = np.zeros(img_shape)
V = np.zeros(img_shape)
f = 335
w = np.array([0, 0, 5]).reshape(3, 1) / 1000
```

```python
for x in range(U.shape[1]):
    for y in range(U.shape[0]):
        xi = x-90
        yi = y-90
#         if np.mod(x, 10) != 0 or np.mod(y, 10) != 0:
#             continue
        m = np.array([
            [(xi*yi)/f, -(f+xi**2 / f), yi],
            [f + yi**2 / f, -xi*yi/f, -xi]
        ])
        r = m @ w
        U[y, x] = r[0]# if not np.isclose(r[0], 0.0) else 0.0
        V[y, x] = r[1] #if not np.isclose(r[1], 0.0) else 0.0
```

```python
plt.figure(figsize=(10, 10))
plt.quiver(U[::10, ::10], V[::10, ::10], cmap="jet_r", angles="xy")
# plt.gca().invert_yaxis()
```

```python
m = plt.imshow(U)
plt.colorbar(m)
```
